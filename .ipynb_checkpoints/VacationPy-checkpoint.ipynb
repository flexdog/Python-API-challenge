```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
#Load .csv file
weather_df=pd.read_csv('weather_df.csv')
weather_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Setup gmaps to display data from the weather notebook of part 1 and use humidity as the weight factor for the heatmap.
gmaps.configure(api_key=g_key)

locations = weather_df[['Lat', 'Lng']]
weights = weather_df['Humidity']

fig=gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(locations, weights=weights))
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Create a desired weather dataframe based on a temperture range of 70 to 80 degrees and a humidity of 30 to 61 percent. 
# We further refine the search with no clouds and a wind speed of less than 10 mph.
desired_weather_df = weather_df[weather_df['Temp'].gt(69) & weather_df['Temp'].lt(81) & weather_df['Wind Speed'].lt(10) & weather_df['Cloudiness'].eq(0)].reset_index()
desired_weather_df.head(10)
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
dWeather_df = desired_weather_df[['City', 'Country', 'Lat', 'Lng']].copy()
dWeather_df['Hotel'] = dWeather_df.apply(lambda _: '', axis=1) 

dWeather_df.head()
```

```python
# Code to create a list for hotels that match our refined search for the "perfect" vacation location.
hotel_list=[]
Lat_list=[]
Lng_list=[]
City_list=[]
Country_list=[]

for row in dWeather_df.itertuples():
    #base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"+"location=" + str(row.Lat) +"," +str(row.Lng) + "&radius=5000&type=hotel&keyword=hotel&key=" + g_key
    
    # Execute request
    response = requests.get(base_url)
    try:
        # Convert response to json
        jaysponse = response.json()
        if jaysponse['status'] == "ZERO_RESULTS":
           response.raise_for_status()
        else:
            print(f"Hotel {jaysponse['results'][0]['name']} found nearby in the country of {row[2]}")
            hotel_list.append(jaysponse['results'][0]['name'])
            Lat_list.append(jaysponse['results'][0]['geometry']['location']['lat'])
            Lng_list.append(jaysponse['results'][0]['geometry']['location']['lng'])
            City_list.append(row[1])
            Country_list.append(row[2])
    except KeyError:
        print(f"No hotel found nearby.")
 
```

```python
#create dataframe
dHotel_dict = {
    "Hotel Name":hotel_list,
    "Lat":Lat_list,
    "Lng":Lng_list,
    "City":City_list,
    "Country":Country_list
}

dHotel_df=pd.DataFrame(dHotel_dict)
dHotel_df.head()
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in dHotel_df.iterrows()]
locations = dHotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers=gmaps.marker_layer(locations, hotel_info)
fig.add_layer(markers)
fig
# Display Map
```
